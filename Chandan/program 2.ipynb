```python
import pandas as pd 
import numpy as np
import statsmodels.formula.api as smf
import seaborn as sns
df=pd.read_csv('pgm2.csv',index_col=0)
df.head()
x=df.lstat.values
y=df.medv.values
print (x)
print(y)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.1,test_size=0.3,random_state=100)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
lm = smf.ols(formula='medv ~ lstat', data=df)
result = lm.fit()
result.rsquared, result.fvalue, result.params.Intercept, result.params.lstat
new = pd.DataFrame([[1, 5], [1, 10], [1, 15]], columns=['Intercept', 'lstat'])
result.predict(new)
from sklearn.linear_model import LinearRegression
linear=LinearRegression()
linear.fit(x_train.reshape(-1,1),y_train)
y_pred_test=linear.predict(x_test.reshape(-1,1))
y_pred_train=linear.predict(x_train.reshape(-1,1))
sns.regplot('lstat', 'medv', df, line_kws = {"color":"r"}, ci=None)
```
